```python
from caracteristicas import sift, match
import cv2
```

```python
imagen1 = cv2.imread('sift1.jpg')
imagen2 = cv2.imread('sift2.jpg')
escalas = 5  
sigma = 2
octavas = 4 
threshold = 50

keypoints1, descriptores1 = sift(imagen1, octavas, escalas, sigma,threshold)
print("Prueba 1 terminada")
keypoints2, descriptores2 = sift(imagen2, octavas, escalas, sigma,threshold)
print("Prueba 2 terminada")

match(imagen1,descriptores1, keypoints1, imagen2,descriptores2,keypoints2, umbral=0.5,matches_thickness=2)
```
